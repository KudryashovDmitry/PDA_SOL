```python
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt
```

```python
def equation(x):
    return (1 + x)**3 - 2*x**2 - 28*x - 17

x0 = [0]
solution = root(equation, x0)
x_solution = solution.x[0]
print(f"x = {x_solution}")
```

```python
x_vals = np.linspace(-5, 5, 100)
y_vals = equation(x_vals)

plt.plot(x_vals, y_vals, label='(1 + x)^3 - 2(x^2) - 28x - 17')
plt.axhline(0, color='r', linestyle='--', label='Уравнение = 0')
plt.scatter(x_solution, 0, color='b', marker='o', label='Решение')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графическое решение уравнения')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def polynomial(x):
    return 28*x**3 - (3 + 1)*x**2 - 1*x + 1

x0 = [-2, 0, 2]
roots = fsolve(polynomial, x0)
print("Корни полинома:")
print(roots)
x_vals = np.linspace(-4, 4, 100)
y_vals = polynomial(x_vals)
plt.plot(x_vals, y_vals, label='28x^3 - (3 + 1)x^2 - 1x + 1')
plt.axhline(0, color='r', linestyle='--', label='Полином = 0')
plt.scatter(roots, np.zeros_like(roots), color='b', marker='o', label='Корни')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графическое решение полинома')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(z):
    x1, x2, x3, x4 = z
    eq1 = 3*x1 + x2 + x3 + 3*x4 - 28
    eq2 = x1 - x2 + x3 + 4*x4 - 7
    eq3 = -5*x1 - x3 - 7*x4 + 5
    eq4 = x1 - 6*x2 + 28*x3 + 6*x4 - 3
    return [eq1, eq2, eq3, eq4]

x0 = [0, 0, 0, 0]
solution = fsolve(equations, x0)
print("Решение системы линейных уравнений:")
print("x1 =", solution[0])
print("x2 =", solution[1])
print("x3 =", solution[2])
print("x4 =", solution[3])
```

```python
from scipy.optimize import root

def equations(v):
    x1, x2, x3, x4 = v
    eq1 = 7*x1 + 3*x2 + 1*x3 + x4 - 11
    eq2 = x1 - 3*x2 + 2*x3 + 112*x4 - 12
    eq3 = -5*x1 + 1*x2 - 1*x3 - 7*x4 + 5
    eq4 = 1*x1 - 6*x2 + 2*x3 + 6*x4 - 1
    return [eq1, eq2, eq3, eq4]

x0 = [0, 0, 0, 0]
solution = root(equations, x0, method='hybr')
print("Решение системы линейных уравнений:")
print("x1 =", solution.x[0])
print("x2 =", solution.x[1])
print("x3 =", solution.x[2])
print("x4 =", solution.x[3])
```

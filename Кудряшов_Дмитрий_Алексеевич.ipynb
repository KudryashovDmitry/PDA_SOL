### Step 3. Assign it to a variable called drinks.

```python
columns='country,beer_servings,spirit_servings,wine_servings,total_litres_of_pure_alcohol,continent'.split(',')
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby('continent').beer_servings.mean().sort_values(ascending=False)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
grouped = drinks.groupby('continent')['wine_servings']

for continent, data in grouped:
    print(f"{continent}       ")
    print(data.describe())
    print()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continent').mean(numeric_only=True)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
Df_1 = drinks.groupby('continent').median(numeric_only=True)
Df_1.to_csv('Кудряшов Дмитрий Алексеевич.csv', index = False)
from google.colab import files
files.download('Кудряшов Дмитрий Алексеевич.csv')
a = pd.read_csv('/content/Кудряшов Дмитрий Алексеевич.csv')
a.head()
```
